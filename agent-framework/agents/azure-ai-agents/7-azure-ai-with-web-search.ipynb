# 📈 Azure AI Agent with Web Search - Financial Market Research 🏦

This notebook demonstrates how to create an Azure AI agent using the Microsoft Agent Framework with the **Web Search tool** for real-time market research, including stock news, economic trends, and financial news.

## Features Covered:
- Setting up Web Search integration (replaces deprecated Bing Grounding)
- Using web search for real-time financial market information
- Creating agents that can access current market news
- Proper resource management with async context managers

### What Changed: Bing Grounding → Web Search
The `bing_grounding` tool type is replaced by `web_search`. No connection ID required — just add the tool definition.

## Import Libraries

Import the required libraries using `AzureAIProjectAgentProvider` and `WebSearchTool`:

````python
import os
from pathlib import Path

from agent_framework_foundry import FoundryChatClient
from azure.ai.projects.models import WebSearchTool
from azure.identity import AzureCliCredential
from dotenv import load_dotenv
````

## Initial Setup

Load environment variables from the `.env` file:

````python
# Map env var name to what agent-framework expects
if os.getenv("AI_FOUNDRY_PROJECT_ENDPOINT") and not os.getenv("AZURE_AI_PROJECT_ENDPOINT"):
    os.environ["AZURE_AI_PROJECT_ENDPOINT"] = os.environ["AI_FOUNDRY_PROJECT_ENDPOINT"]

# Load environment variables from .env file (three directories up)
load_dotenv('../../.env')

# Verify environment setup
endpoint = os.getenv('AI_FOUNDRY_PROJECT_ENDPOINT')
model = os.getenv('FOUNDRY_MODEL')

print("🔧 Environment Configuration:")
print(f"✅ Project Endpoint: {endpoint[:50]}..." if endpoint else "❌ AI_FOUNDRY_PROJECT_ENDPOINT not set")
print(f"✅ Model Deployment: {model}" if model else "❌ FOUNDRY_MODEL not set")
````

## Configure Web Search Tool 🌐

Define the web search tool using the dict-based format supported by `AzureAIProjectAgentProvider`:

````python
# Web Search tool configuration — no connection ID required
web_search_tool = WebSearchTool()

print("✅ Web Search tool configured")
````

## Create and Run the Financial Market Research Agent 📈

Create an Azure AI agent with web search for financial market research:

````python
async def main() -> None:
    """Demonstrate Azure AI agent with Web Search for financial market research."""
    
    credential = AzureCliCredential(process_timeout=30)
    client = FoundryChatClient(credential=credential)
    agent = client.as_agent(
        name="FinancialMarketResearchAgent",
        instructions=(
            "You are a Financial Market Research Analyst with web search capabilities. "
            "Search for current financial news, stock market updates, economic trends, and investment insights. "
            "Always cite your sources and include relevant dates. "
            "Remind users that past performance doesn't guarantee future results."
        ),
        tools=[web_search_tool],
    )
    
    async with agent:
        print("=== 📈 Azure AI Financial Market Research Agent ===\n")
        
        user_input = "What are the latest trends in the banking and financial services sector?"
        print(f"🤔 Analyst Query: {user_input}")
        response = await agent.run(user_input)
        print(f"📊 Market Research: {response.text}\n")
        print("⚠️ Disclaimer: This is for research purposes only. Consult a financial advisor for investment decisions.")
````

## Execute the Example 🚀

Run the main function:

````python
await main()
````

## Additional Market Research Queries 📊

Test the agent with more FSI research queries:

````python
async def financial_market_research_queries():
    """Test the agent with multiple FSI market research queries."""
    
    credential = AzureCliCredential(process_timeout=30)
    client = FoundryChatClient(credential=credential)
    agent = client.as_agent(
        name="FinancialMarketResearchAgent",
        instructions=(
            "You are a Financial Market Research Analyst specializing in banking and financial services. "
            "Search for current market information, economic indicators, and industry trends. "
            "Always cite sources, include dates, and note this is for informational purposes only."
        ),
        tools=[web_search_tool],
    )
    
    async with agent:
        queries = [
            "What are the current Federal Reserve interest rate expectations?",
            "What are the latest news about major bank earnings and financial sector performance?",
            "What is the current state of the mortgage market and housing finance trends?",
        ]
        
        for query in queries:
            print(f"\n{'='*60}")
            print(f"🔍 Research Query: {query}")
            print("-"*40)
            response = await agent.run(query)
            print(f"📊 Analysis: {response.text}")
            print("⚠️ For informational purposes only.")

await financial_market_research_queries()
````
